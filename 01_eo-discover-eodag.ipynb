# Template Notebook for TUW JupyterHub

The TUW JupyterHub is pre-configured with harmonized EO data access. EO data can be queried with the EODAG client. 

```python
from eodag.api.core import EODataAccessGateway
dag = EODataAccessGateway()
```

EO data from a specific product type, like Sentinel 1 Ground Range Detected backscattering (`S1_SAR_GRD`), can be filtered by a temporal and spatial extent, like so:

```python
product_type = 'S1_SAR_GRD'
footprint = {'lonmin': 0, 'latmin': 43.75, 'lonmax': 0.75, 'latmax': 44}
products, _ = dag.search(productType=product_type,
                         geom=footprint,
                         start='2018-01-20',
                         end='2018-01-23',
                         items_per_page=3)
products
```

Following data discovery, the products can be downloaded, like so:

```python
dag.download_all(products)
```
